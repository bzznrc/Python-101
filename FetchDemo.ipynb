# Grouping Stuff ü•Ç
How nice it is when our data come exactly in the format we need them into?
Not sure, never happened to me. ü§∑‚Äç‚ôÇÔ∏è

Most likely, when you have a DataFrame you'll want to produce a few additional ones with different kinds of grouping, in order to see suff from different perspectives.

For example, with starting from the original DataFram we could ask:
- How many entries do we have for each category_3?
- How many different descriptions do we have for each category_3?

In order to answer to both questions, we'll definitely need to **group** the data by *category_3*.

Grouping the data by a field (or more than one fields) means producing a new DataFrame in which each row will correspond to only one grouping values.

Let's see an example:

```python
#First, we group the df by 'category_3'
#Then we consider only the 'description' column
#On that column, we count how many occurrences we have (for each group)
df.groupby(by='category_3')['description'].count()
```

```python
#First, we group the df by 'category_3'
#Then we consider only the 'description' column
#On that column, we count how many distinct values we have (for each group)
df.groupby(by='category_3')['description'].nunique()
```

In the context of this excercise, we'll use Plotly Express, which is probably the simplest out there.

```python
#First we import the library like this:
import plotly.express as px

#This is the result from the previous excercise. Let's save it in a DataFrame
df_cat = df.groupby(by='category_3')['description'].nunique()

# 1 - Then, we'll use the histogram function to plot it right away
#px.bar(df_cat)

# 2- Cool. let's try to sort the values before plotting
#px.bar(df_cat.sort_values(ascending=False))

# 3 - Super cool. let's try to make a few cosmetic changes
#The labels parameter lets you use a Dictionary to rename the axis. The update_layout method lets you remove the legend
px.bar(df_cat.sort_values(ascending=False), labels={'value':'Count of SKUs', 'category_3':'Product Category'}).update_layout(showlegend=False)
```
