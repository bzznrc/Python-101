# Program
Programming a computer to solve a problem always requires defining **data** to represent the things in your problem; this is called modeling. It also requires explaining **unambiguously** what process to follow in order to solve the problem.

In Imperative programming (ours and the most well-known), we tell the computer what to do to the data. The computer starts out with its memory containing stuff, we tell it how to change that stuff, and it makes those changes.

A program is ran, **one line at a time, from top to bottom**.

Usually you can identify:

> **Inputs** - Amount in Lire / Letters for Stickman

> **Logic** - Calculating conversion / Checking if you guessed the word 

> **Outputs** - Euro value / Whether you've won or lost




```
### 1 - INPUT - Asking for data
lire = input("Inserisci LIRE: ")  

### 2 - LOGIC (is usually quite longer)
eur = float (lire) /1936.27

### 3 - OUTPUT - Communicating the result
print("EUR:", eur)
```

#Cast
Lavorando con la funzione input() incontreremo presto un problema abbastanza spiacevole.
Nel primo esempio qua sotto, chiedo un numero all'utente (le lire) e cerco di calcolare gli euro, dividendo per 1936.27.

L'errore deriva dal fatto che, di default, input() restituisce **testo**.
Se, ad esempio io inserisco 40000 (quarantamila lire), quello che finirá nella variabile lire sará "40000". Ovvero una stringa di testo che contiene il valore testuale "40000". A questo punto cercheró di dividere un testo per 1936.27, cosa che genererá l'errore.

La funzione **float()** serve a dire a Python che il valore che l'utente ha inserito é da considerarsi come un numero con la virgola, invece che come una stringa di testo.
Posso fare la stessa cosa trasformando la variabile in un numero intero con la funzione **int()**.

```
#Errore
#Variabile testuale: "40000"
#Variabile numerica: 40000
lire = input("Inserisci LIRE: ")
eur = lire / 1936.27
print("EUR:", eur)
```

```
#No errore
lire = input("Inserisci LIRE: ")
eur = float (lire) /1936.27
print("EUR:", eur)
```
